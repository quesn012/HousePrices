My guess is these homes are just missing garages or basements. This wouldn't really be an NA value but more of a new category, something along the lines of 'missing' so in this case I will fill the NA with a missing feature.

```python
train[['BsmtQual', 'BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2' \
      ,'GarageType','GarageYrBlt','GarageFinish','GarageQual','GarageCond']] = \
      train[['BsmtQual', 'BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2' \
      ,'GarageType','GarageYrBlt','GarageFinish','GarageQual','GarageCond']].fillna(value='Missing')
```

```python
missing_col = train.isnull().any()
missing_col = missing_col[missing_col==True]
data_missing_col = train[missing_col.index]
data_missing_col.describe(include = 'all')
```

We still have a few missing values to deal with. Quite a few of them don't seem super relevant and are VERY sparse. MiscFeature, PoolQC, Alley and Fence are all sparse enough where I don't think they would have great predictive power.

That still leaves LotFrontage, MasVnrType, MasVnrArea, Electrical, FireplaceQu. 

For the numericals (LotFrontage and MasVnrArea) I'll make the assumption these are missing and should be 0 (no lot or masVnr area exists).

For the categoricals (Electrical, MasVnrType, FirePlaceQu) I'll fill them in a similar way to basement/garage above with a 'none' categorical. I'm making the assumption here there is no fire place/no MasVnrType and that the electrical is... very basic? Not exactly sure about that one, but it is only a single missing value so it shouldn't make much of a difference regardless.

```python
train[['LotFrontage','MasVnrType']] = \
      train[['LotFrontage','MasVnrType']].fillna(value=0)
```

```python
train[['Electrical','MasVnrType','FireplaceQu']] = \
      train[['Electrical','MasVnrType','FireplaceQu']].fillna(value="None")
```
